```python
# train_data 확인 2022-01-01 ~ 2023-04-04
data = pd.read_csv('drive/MyDrive/LG AIMERS/최연아/open/train.csv')
```

```python
# '2022-04-25' 열의 인덱스를 찾기
column_to_stop = "2022-04-25"
stop_index = data.columns.get_loc(column_to_stop)

# '2022-04-25' 열까지만 유지하고 나머지를 제거하기
data = data.iloc[:, :stop_index + 1]
data.head()
```

```python
# '2022-04-05' 열의 위치 찾기
column_index = data.columns.get_loc('2022-04-05')

# '2022-04-25' 열과 그 이후의 모든 열 선택
data = data.iloc[:, column_index:]

# 결과 확인
data.head()
```
