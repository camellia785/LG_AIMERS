```python
# train_data 확인 2022-01-01 ~ 2023-04-04

train_data = pd.read_csv('drive/MyDrive/LG AIMERS/최연아/open/train.csv').drop(columns=['ID', '제품'])
train_data.head()
```

```python
sales_data = pd.read_csv('drive/MyDrive/LG AIMERS/최연아/open/sales.csv').drop(columns=['ID', '제품'])
sales_data.head()
```

```python
 #row는 하나의 날짜, 하나의 대/중/소분류에 대한 하나의 판매량
train_melted = pd.melt(train_data, id_vars=["대분류", "중분류", "소분류", "브랜드"], var_name="날짜", value_name="판매량")
sales_melted = pd.melt(sales_data, id_vars=["대분류", "중분류", "소분류", "브랜드"], var_name="날짜", value_name="판매금액")
```

```python
# 날짜별로 대/중/소분류의 판매량과 판매금액
train_grouped = train_melted.groupby(["날짜", "대분류", "중분류", "소분류"]).sum().reset_index()
sales_grouped = sales_melted.groupby(["날짜", "대분류", "중분류", "소분류"]).sum().reset_index()
```

```python
# 대분류, 중분류, 소분류별로 train 데이터를 그룹화하고 합계를 계산
train_grouped = train_data.groupby(['대분류', '중분류', '소분류']).sum()

# 대분류, 중분류, 소분류별로 sales 데이터를 그룹화하고 합계를 계산
sales_grouped = sales_data.groupby(['대분류', '중분류', '소분류']).sum()

# train과 sales 데이터를 병합
merged_data = pd.merge(train_grouped.reset_index(), sales_grouped.reset_index(), on=['대분류', '중분류', '소분류'], suffixes=('_train', '_sales'))

# 대분류별 상관관계 계산
corr_by_big_category = merged_data.groupby('대분류').apply(lambda x: x.loc[:, x.columns.str.endswith('_train')].corrwith(x.loc[:, x.columns.str.endswith('_sales')]))

# 중분류별 상관관계 계산
corr_by_medium_category = merged_data.groupby('중분류').apply(lambda x: x.loc[:, x.columns.str.endswith('_train')].corrwith(x.loc[:, x.columns.str.endswith('_sales')]))

# 소분류별 상관관계 계산
corr_by_small_category = merged_data.groupby('소분류').apply(lambda x: x.loc[:, x.columns.str.endswith('_train')].corrwith(x.loc[:, x.columns.str.endswith('_sales')]))

print("대분류별 상관관계:", corr_by_big_category)
print("중분류별 상관관계:", corr_by_medium_category)
print("소분류별 상관관계:", corr_by_small_category)
```
